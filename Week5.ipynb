```python
import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import glob
import math
```

Gamma Correction

```python
print(' Gamma Correction pada citra ')
print('----------------------------------')
try:
  gamma = int(input('masukkan nilai gamma : '))
except ValueError:
  print('Error, not a number')

image = cv.imread('/content/drive/MyDrive/Asset/Images/female.tiff')
image_gamma = np.zeros(image.shape, image.dtype)

for y in range(image.shape[0]):
  for x in range(image.shape[1]):
    for c in range(image.shape[2]):
      image_gamma[y,x,c] = np.clip(255 * pow(image[y,x,c]/255,(1/gamma)), 0,255)

final_frame = cv.hconcat((image, image_gamma))
cv2_imshow(final_frame)
```

Image Depth

```python
print(' Image Depth pada citra : ')
print('-------------------------------------')
try: 
 depth = int(input('Masukkan nilai image depth  : '))
except ValueError:
 print('Error, not a number')


image = cv.imread('/content/drive/MyDrive/Asset/Images/jungle.png')
depth_image = np.zeros(image.shape, image.dtype)

akurasi = 255/((2**depth)-1)

for y in range(image.shape[0]):
  for x in range(image.shape[1]):
    for c in range(image.shape[2]):
      depth_image[y,x,c] = np.clip( round((image[y,x,c]/akurasi))*akurasi, 0,255)


final_frame = cv.hconcat((image, depth_image))
cv2_imshow(final_frame)
```

Average Denoising

```python
#Jumlah Citra = 5

import glob
from math import log10,sqrt

original = cv.imread('/content/drive/MyDrive/Asset/Images/galaxy.jpg')

cv_img = []
for img in glob.glob('/content/drive/MyDrive/Asset/Images/noises/*.jpg'):
  n= cv.imread(img)
  cv_img.append(n)

noise_img = cv_img[4]


for i in range(4):
  alpha = 1.0 / (i + 1)
  beta = 1.0 - alpha
  noise_img = cv.addWeighted(cv_img[i], alpha, noise_img, beta, 0.0)

def PSNR(original, compressed): 
    mse = np.mean((original - compressed) ** 2) 
    if(mse == 0): # MSE is zero means no noise is present in the signal. 
                  # Therefore PSNR have no importance.
        return 100
    max_pixel = 255.0
    psnr = 20 * log10(max_pixel / sqrt(mse)) 
    return psnr 
  
compressed = noise_img
psnr = PSNR(original, compressed)

print('---------------------------')
print('Jumlah Average Citra = 5')
print('Nilai PSNR =', psnr, 'dB')

final_frame = cv.hconcat((original, noise_img))
cv2_imshow(final_frame)
```
